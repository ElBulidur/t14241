analise de notas de alunos

```python
alunos = ['Ana', 'Jefesson', 'Lucas', 'Paulo', 'Talita', 'Meggy']
disciplinas = ['series', 'dataframe', 'plotagem', 'machine learning']

notas = [] # [(0,1),(0,2)]

# adicionando notas de 0 a 10 aleatorio aos alunos de acordo com as disciplinas
# crie uma lista de tuplas contendo o index do aluno e sua notas referente a sua disciplina
# from random import randint
from numpy.random import randint # Gera numeros inteiros aleatoria

for aluno in alunos: #
    for disciplina in disciplinas:
        notas.append((alunos.index(aluno), randint(0,10))) # (0, 0)

notas
```

```python
# adicionando notas de 0 a 10 aleatorio aos alunos de acordo com as disciplinas
# Criei um dicionario com os nomes dos alunos e suas disciplinas com notas
alunos_com_notas = {}

for aluno in alunos:
    alunos_com_notas[aluno] = []
    
    for disciplina in disciplinas:
        alunos_com_notas[aluno].append({disciplina:randint(0,10)})
    
alunos_com_notas['Ana']
```

```python
# Criando series de alunos
s_alunos = pd.Series(alunos_com_notas)
s_alunos
```

```python
# Criando um dataframe
df_alunos = pd.DataFrame(alunos_com_notas)
df_alunos
```
