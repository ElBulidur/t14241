```python
# importacoes

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data2.csv')

df.head()
df.shape
df.columns
```

```python
df1 = df.groupby(['Name'])[['Pulse', 'Maxpulse']].agg([np.mean]).reset_index()

fig = plt.figure(figsize=(12,5))

plt.plot(df1['Name'], df1['Pulse'])

plt.xlabel('Nomes dos corredores')
plt.ylabel('Pulsacao por corredor')

plt.grid(visible=True)
```

```python
df2 = pd.read_csv('data2.csv')
df2.shape
df2.columns
df2.head()
```

```python
fig = plt.figure(figsize=(12,5))

plt.plot(df2.index, df2.Calories)

plt.xlabel('Todas as corridas')
plt.ylabel('Calorias')

plt.grid(visible=True)
```

```python
fig = plt.figure(figsize=(12,5))

plt.bar(df2.index, df2.Calories)

plt.xlabel('Todas as corridas')
plt.ylabel('Calorias')

plt.grid(visible=True)
```

```python
fig = plt.figure(figsize=(12,5))

plt.hist(df2.Calories, bins =100)

plt.xlabel('Frequencia')
plt.ylabel('Calorias - Distribuida por frequencia')

plt.grid(visible=True)
```

```python
import scipy.stats as stats

fig = plt.figure(figsize=(16,5))

x = df.Pulse.sort_values()

med = df.Pulse.mean()

sd = df.Pulse.std()

plt.plot(x, stats.norm.pdf(x, med,sd))
```

```python
fig = plt.figure(figsize=(16,5))

x = df.Pulse.sort_values() # variavel a analisar

med = df.Pulse.mean() # media da variavel

sd = df.Pulse.std() # devio-padrao da variavel

freq = x.value_counts()

freq_m = freq.max()

dens = stats.norm.pdf(x, med, sd)

dens_m = dens.max()

fator = freq_m / dens_m

plt.plot(x, dens * fator)
plt.hist(df.Pulse, bins=100)

plt.ylabel('Frequencia')
plt.xlabel('Freq. Cardiaca')

plt.grid(visible=True)
```
