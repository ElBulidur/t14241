```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
import pandas as pd
from numpy.random import randint

alunos = ['Julio','Ricardo','Talita','Meggy']

notas = {
    'alunos': alunos,
    '1bim': [randint(11) for _ in range(len(alunos))],
    '2bim': [randint(11) for _ in range(len(alunos))],
    '3bim': [randint(11) for _ in range(len(alunos))],
    '4bim': [randint(11) for _ in range(len(alunos))],
}

df1 = pd.DataFrame(data=notas)
df1.head()
```

# plot()

```python
# Anatomia matplotlib => https://matplotlib.org/stable/gallery/showcase/anatomy.html
# markers => https://matplotlib.org/stable/api/markers_api.html?highlight=maplotlib+axes+marker
# LineStyles = https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html

ax1 = df1.plot(\
    title='Notas de alunos', legend='xxx', grid=True, fontsize=14,\
    xlabel='ALunos', ylabel='Notas', kind='line', \
    ylim=(0,10), xlim=(0,3), colormap='plasma',\
    # deixar a apresentacao mais detalhada
    subplots=True, layout=(2,2), figsize=(14,3),\
    #table=df1['1bim']
    marker='o', markersize=4, mfc='c', mec='g', linestyle='--', \
    xticks=[0,1,2,3], rot=90
    )
```

```python
df2 = df1.drop(columns='alunos')
df2.index = alunos
print(df1.head())
df2.head()
```

```python
ax2 = df2.plot.line(marker='o', colormap='Dark2', rot=90)

ax2.set_title('Notas dos alunos', fontsize=26, color='r')

# legendas alteradas
handles, labels = ax2.get_legend_handles_labels()

nova_legenda = ['1 Bimestre', '2 Bimestre', '3 Bimestre', '4 Bimestre']

ax2.legend(nova_legenda, bbox_to_anchor=(1.26,1), loc=0, borderaxespad=0)

ax2.set_xlabel('Alunos', color='c')
ax2.set_ylabel('Notas', color='c')
ax2.set_xlim(0,3)
ax2.set_xticks([0,1,2,3])
ax2.set_xticklabels(alunos, color='grey')
ax2.set_facecolor('aquamarine')

texto = 'explicacao sobre \ngrafico referente \nas notas dos alunos'

box = {'boxstyle': 'round', 'facecolor': 'yellow', 'alpha': 0.9}

ax2.text(0.64,0.75,texto, transform=ax2.transAxes, fontsize=10, color='g', bbox=box)

# Funcao para colocar valores no grafico
def setar_valores(df):
    for i,valor in enumerate(df, start=0):
        ax2.text(x=df[i], y=valor, s=f'({valor})', position=(i,valor), ha='left', color='gray')


# chamando funcoes para as variaveis
setar_valores(df2['1bim'])
setar_valores(df2['2bim'])
setar_valores(df2['3bim'])
setar_valores(df2['4bim'])
```
